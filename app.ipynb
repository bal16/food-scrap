```python
url = "https://cookpad.com/id/cari/makanan%20sehat"
url
```

```python
page = requests.get(url).text
page
```

```python
page_soup = bs4(page, "html.parser")
page_content = page_soup.find_all("li", 
    class_= "block-link card border-cookpad-gray-400 border-t-0 border-l-0 border-r-0 border-b flex lg:flex-row-reverse m-0 rounded-none overflow-hidden ranked-list__item xs:border-b-none xs:mb-sm xs:rounded-lg hover:bg-cookpad-gray-200 print:bg-cookpad-white")
```

```python
food_names = []
food_images = []
food_recipes = []
```

```python
for food in page_content:
    food_soup = bs4(str(food), "html.parser")
    name = food_soup.find("h2">"a", {"class": "block-link__main"}).get_text(strip=True)
    image = food_soup.find("picture">"img", {"class":"h-full object-cover"}).get("src")
    recipes= food_soup.find("div", {"data-ingredients-highlighter-target":"ingredients"}).get_text(strip=True)
    recipes = recipes.split(",\n      ")
    food_names.append(name)
    food_images.append(image)
    food_recipes.append(recipes)
```

```python
data = {
    "name": food_names,
    "image": food_images,
    "recipe": food_recipes
}
```

```python
df = pd.DataFrame(data)
df.to_csv("output/food.csv", index=False)

df
```
